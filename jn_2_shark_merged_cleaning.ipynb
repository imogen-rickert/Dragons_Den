## Examine and work on missing values

### Drop rows with too many missing values

All the columns from 1 dataset have the same missing 103 values (and some have a few more missing values) that the other dataset has values for. We decided that it is not useful to have rows with only a few values so we removed them.

```python
missing_values = shark_merged_copy.isna().sum().sort_values(ascending = False)
missing_values
```

```python
null_display = shark_merged_copy[(shark_merged_copy['season'].isnull()==True)]
null_display = null_display[['season', 'shark1' , 'pitched_business_desc', 'pitched_business_identifier', 'season_epi_code']]
null_display.head(10)
```

```python
shark_merged_copy = shark_merged_copy.dropna(subset=['shark1', 'shark2', 'shark3'])
```

```python
shark_merged_copy.shape
```

```python
missing_values2 = shark_merged_copy.isna().sum().sort_values(ascending = False)
missing_values2
```

### Replace null values in 'deal_shark' column with 'Unknown'

Looking at this column, the null values likely indicate that we do not know which shark(s) made the deal, so we can change this to 'Unknown'

```python
shark_merged_copy.head()
```

```python
print(set(shark_merged_copy['deal_shark']))
```

```python
shark_merged_copy["deal_shark"].fillna("Unknown", inplace = True) 
```

```python
print(set(shark_merged_copy['deal_shark']))
```

#### End result of cleaning missing values

```python
missing_values3 = shark_merged_copy.isna().sum().sort_values(ascending = False)
missing_values3
```
